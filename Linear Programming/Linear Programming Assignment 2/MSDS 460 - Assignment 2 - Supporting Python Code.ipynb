#### Answer

```python
# Let's solve this problem using the simplex method

# Import necessary methods and functions
from pulp import LpVariable, LpProblem, LpMaximize, LpStatus, value, LpMinimize
import pandas as pd

# Create linear programming variables
Cars = LpVariable("Cars", 0, None) # Cars >= 0 & no upper bound
Trucks = LpVariable("Trucks", 0, None) # Trucks >= 0 & no upper bound
Machine_Ones = LpVariable("Machine_Ones", 0, None) # Machine_Ones >= 0 & no upper bound 

# Create the problem variable
prob2 = LpProblem("problem", LpMaximize)

# Define the constraints - remember that the lower bounds are defined in LpVariable above

######## Daily Napkin Demand Constraints
prob2 += (2 * Cars) + (3 * Trucks) <= 260                    # Steel Production Constraint
prob2 += (0.6 * Cars) + ( 0.7 * Trucks) <= 73                # Machine 2 Production Constraint
prob2 += (0.8 * Cars) +  Trucks - Machine_Ones <= 0          # Machine 1 Production Constraint
prob2 += Machine_Ones <= 98                                  # Constraint On Supply of Machine 1's For Rent
prob2 += Cars >= 88                                          # Constraint For Car Demand
prob2 += Trucks >= 26                                        # Constraint For Truck Demand

# Define the objective function 
prob2 += ( 300 * Cars) + (400 * Trucks) - (50 * Machine_Ones) 

# Solve the problem
status = prob2.solve()
LpStatus[status]

# Print the solution
print ("Optimal Solution:")
for variable in prob2.variables():
    print(variable.name, " = ", value(variable))

print("Total Profit = $", round(value(prob2.objective),2))
```

```python
# Let's solve this problem using the simplex method

# Import necessary methods and functions
from pulp import LpVariable, LpProblem, LpMaximize, LpStatus, value, LpMinimize
import pandas as pd

# Create linear programming variables
Prof1_Fall_Marketing = LpVariable("Prof1_Fall_Marketing", 0, None) # Prof1_Fall_Marketing >= 0 & no upper bound
Prof1_Fall_Finance = LpVariable("Prof1_Fall_Finance", 0, None) # Prof1_Fall_Finance >= 0 & no upper bound
Prof1_Fall_Production = LpVariable("Prof1_Fall_Production", 0, None) # Prof1_Fall_Production >= 0 & no upper bound 
Prof1_Spring_Marketing = LpVariable("Prof1_Spring_Marketing", 0, None) # Prof1_Spring_Marketing >= 0 & no upper bound
Prof1_Spring_Finance = LpVariable("Prof1_Spring_Finance", 0, None) # Prof1_Spring_Finance >= 0 & no upper bound
Prof1_Spring_Production = LpVariable("Prof1_Spring_Production", 0, None) # Prof1_Spring_Production >= 0 & no upper bound
Prof2_Fall_Marketing = LpVariable("Prof2_Fall_Marketing", 0, None) # Prof2_Fall_Marketing >= 0 & no upper bound
Prof2_Fall_Finance = LpVariable("Prof2_Fall_Finance", 0, None) # Prof2_Fall_Finance >= 0 & no upper bound
Prof2_Fall_Production = LpVariable("Prof2_Fall_Production", 0, None) # Prof2_Fall_Production >= 0 & no upper bound 
Prof2_Spring_Marketing = LpVariable("Prof2_Spring_Marketing", 0, None) # Prof2_Spring_Marketing >= 0 & no upper bound
Prof2_Spring_Finance = LpVariable("Prof2_Spring_Finance", 0, None) # Prof2_Spring_Finance >= 0 & no upper bound
Prof2_Spring_Production = LpVariable("Prof2_Spring_Production", 0, None) # Prof2_Spring_Production >= 0 & no upper bound
Prof3_Fall_Marketing = LpVariable("Prof3_Fall_Marketing", 0, None) # Prof3_Fall_Marketing >= 0 & no upper bound
Prof3_Fall_Finance = LpVariable("Prof3_Fall_Finance", 0, None) # Prof3_Fall_Finance >= 0 & no upper bound
Prof3_Fall_Production = LpVariable("Prof3_Fall_Production", 0, None) # Prof3_Fall_Production >= 0 & no upper bound 
Prof3_Spring_Marketing = LpVariable("Prof3_Spring_Marketing", 0, None) # Prof3_Spring_Marketing >= 0 & no upper bound
Prof3_Spring_Finance = LpVariable("Prof3_Spring_Finance", 0, None) # Prof3_Spring_Finance >= 0 & no upper bound
Prof3_Spring_Production = LpVariable("Prof3_Spring_Production", 0, None) # Prof3_Spring_Production >= 0 & no upper bound

# Create the problem variable
prob = LpProblem("problem", LpMaximize)

# Define the constraints - remember that the lower bounds are defined in LpVariable above

######## Constraints That Professors Teach 4 Courses Per Year
prob += Prof1_Fall_Marketing + Prof1_Fall_Finance + Prof1_Fall_Production + Prof1_Spring_Marketing + Prof1_Spring_Finance + Prof1_Spring_Production == 4
prob += Prof2_Fall_Marketing + Prof2_Fall_Finance + Prof2_Fall_Production + Prof2_Spring_Marketing + Prof2_Spring_Finance + Prof2_Spring_Production == 4
prob += Prof3_Fall_Marketing + Prof3_Fall_Finance + Prof3_Fall_Production + Prof3_Spring_Marketing + Prof3_Spring_Finance + Prof3_Spring_Production == 4

######## Constraints That Four Sections of Each Class Must Be Offered Per Year
prob += Prof1_Fall_Marketing + Prof2_Fall_Marketing + Prof3_Fall_Marketing + Prof1_Spring_Marketing + Prof2_Spring_Marketing + Prof3_Spring_Marketing >= 4
prob += Prof1_Fall_Finance + Prof2_Fall_Finance + Prof3_Fall_Finance + Prof1_Spring_Finance + Prof2_Spring_Finance + Prof3_Spring_Finance >= 4
prob += Prof1_Fall_Production + Prof2_Fall_Production + Prof3_Fall_Production + Prof1_Spring_Production + Prof2_Spring_Production + Prof3_Spring_Production >= 4

######## Constraints that One Section of Each Class Must Be Offered In Each Semester:
prob += Prof1_Fall_Marketing + Prof2_Fall_Marketing + Prof3_Fall_Marketing >= 1
prob += Prof1_Fall_Finance + Prof2_Fall_Finance + Prof3_Fall_Finance >= 1
prob += Prof1_Fall_Production + Prof2_Fall_Production + Prof3_Fall_Production >= 1
prob += Prof1_Spring_Marketing + Prof2_Spring_Marketing + Prof3_Spring_Marketing >= 1
prob += Prof1_Spring_Finance + Prof2_Spring_Finance + Prof3_Spring_Finance >= 1
prob += Prof1_Spring_Production + Prof2_Spring_Production + Prof3_Spring_Production >= 1

# Define the objective function to maximize
prob += (9*Prof1_Fall_Marketing) + (8*Prof1_Fall_Finance) + (7*Prof1_Fall_Production) + (10*Prof1_Spring_Marketing) + (9*Prof1_Spring_Finance) + (8*Prof1_Spring_Production) + (9*Prof2_Fall_Marketing) + (11*Prof2_Fall_Finance) + (10*Prof2_Fall_Production) + (7*Prof2_Spring_Marketing) + (9*Prof2_Spring_Finance) + (8*Prof2_Spring_Production) + (9*Prof3_Fall_Marketing) + (8*Prof3_Fall_Finance) + (10*Prof3_Fall_Production) + (9*Prof3_Spring_Marketing) + (8*Prof3_Spring_Finance) + (10*Prof3_Spring_Production)

# Solve the problem
status = prob.solve()
LpStatus[status]

# Print the solution
print ("Optimal Solution:")
for variable in prob.variables():
    print(variable.name, " = ", value(variable), " courses taught by this professor on this subject during this semester")

print("Total Satisfaction = ", value(prob.objective), " units")
```
